```

# Sistema de gestión de tareas usando listas, pilas, colas y árboles

# ----------- LISTA -----------
class ListaTareas:
    def __init__(self):
        self.tareas = []

    def insertar(self, tarea):
        self.tareas.append(tarea)

    def eliminar(self, tarea):
        if tarea in self.tareas:
            self.tareas.remove(tarea)

    def localizar(self, tarea):
        return tarea in self.tareas

    def mostrar(self):
        return self.tareas

# ----------- PILA -----------
class PilaTareas:
    def __init__(self):
        self.pila = []

    def push(self, tarea):
        self.pila.append(tarea)

    def pop(self):
        if not self.esta_vacia():
            return self.pila.pop()

    def peek(self):
        if not self.esta_vacia():
            return self.pila[-1]

    def esta_vacia(self):
        return len(self.pila) == 0

    def mostrar(self):
        return self.pila

# ----------- COLA -----------
class ColaTareas:
    def __init__(self):
        self.cola = []

    def encolar(self, tarea):
        self.cola.append(tarea)

    def desencolar(self):
        if not self.esta_vacia():
            return self.cola.pop(0)

    def primero(self):
        if not self.esta_vacia():
            return self.cola[0]

    def ultimo(self):
        if not self.esta_vacia():
            return self.cola[-1]

    def esta_vacia(self):
        return len(self.cola) == 0

    def mostrar(self):
        return self.cola

# ----------- ÁRBOL BINARIO -----------
class NodoArbol:
    def __init__(self, tarea):
        self.tarea = tarea
        self.izquierda = None
        self.derecha = None

def insertar_arbol(raiz, tarea):
    if raiz is None:
        return NodoArbol(tarea)
    if tarea < raiz.tarea:
        raiz.izquierda = insertar_arbol(raiz.izquierda, tarea)
    else:
        raiz.derecha = insertar_arbol(raiz.derecha, tarea)
    return raiz

def mostrar_arbol(raiz):
    if raiz:
        mostrar_arbol(raiz.izquierda)
        print(raiz.tarea)
        mostrar_arbol(raiz.derecha)

# ------------------ EJEMPLO DE USO ------------------

# Lista de tareas
lista = ListaTareas()
lista.insertar("Tarea 1: Leer")
lista.insertar("Tarea 2: Escribir")
lista.insertar("Tarea 3: Repasar")
print("Lista:", lista.mostrar())

# Pila de tareas
pila = PilaTareas()
pila.push("Tarea urgente A")
pila.push("Tarea urgente B")
print("Pila antes de pop:", pila.mostrar())
print("Pop:", pila.pop())
print("Pila después:", pila.mostrar())

# Cola de tareas
cola = ColaTareas()
cola.encolar("Tarea pendiente X")
cola.encolar("Tarea pendiente Y")
cola.encolar("Tarea pendiente Z")
print("Cola:", cola.mostrar())
print("Primero en la cola:", cola.primero())
print("Último en la cola:", cola.ultimo())
print("Desencolando:", cola.desencolar())
print("Cola después:", cola.mostrar())

# Árbol de tareas
raiz = None
for tarea in ["Tarea M", "Tarea A", "Tarea Z", "Tarea C"]:
    raiz = insertar_arbol(raiz, tarea)
print("Árbol en orden:")
mostrar_arbol(raiz)
```
